### 구매선호지수, 검색선호지수 데이터 불러오기

```python
Online_Preference_Count = pd.read_csv("C:/Users/lsh93/Desktop/data/UBCF2/Online_Preference_Count.csv", encoding="UTF-8")
Online_Preference_Search = pd.read_csv("C:/Users/lsh93/Desktop/data/UBCF2/Online_Preference_Search.csv", encoding="UTF-8")
```

```python
opc = Online_Preference_Count
ops = Online_Preference_Search
```

```python
print(opc.shape)
print(ops.shape)
```

```python
opc = opc.rename(columns = {'Unnamed: 0':'CLNT_ID'})
ops = ops.rename(columns = {'Unnamed: 0':'CLNT_ID'})
```

```python
client1 = opc[['CLNT_ID']]
client2 = ops[['CLNT_ID']]
overlap = pd.merge(client1,client2)
```

```python
print(len(overlap))
overlap.head()
```

```python
opc = pd.merge(opc, overlap)
ops = pd.merge(ops, overlap)
print(len(opc))
print(len(ops))
```

```python
#'CLNT_ID' 빼고 나머지 element끼리 더하기 
sum_op = opc + ops
sum_op = sum_op.drop(sum_op.columns[[0]], axis=1)
sum_op.head()
```

```python
ensemble_op = sum_op/2
ensemble_op.head()
```

```python
ensemble_op = pd.concat([overlap,ensemble_op], axis = 1)
ensemble_op.head()
```
